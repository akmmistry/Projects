```python
# we don't need all those columns - only keep useful ones
cols = ['School Number', 'Board Name',
       'Grade Range', 'Enrolment',
       'Percentage of Students Whose First Language Is Not English',
       'Percentage of Grade 3 Students Achieving the Provincial Standard in Reading',
       'Percentage of Grade 3 Students Achieving the Provincial Standard in Writing',
       'Percentage of Grade 3 Students Achieving the Provincial Standard in Mathematics',
       'Percentage of Children Who Live in Low-Income Households',
       'Percentage of Students Whose Parents Have Some Unviersity Education']
    
df= df[cols]
```

```python
df.rename(columns={'Percentage of Students Whose Parents Have Some Unviersity Education': 'EducatedParents'},inplace=True)
df.rename(columns={'Percentage of Children Who Live in Low-Income Households': 'LowIncomeHome'},inplace=True)
df.rename(columns={'Percentage of Students Whose First Language Is Not English': 'NotEnglish'},inplace=True)

df.rename(columns={'Percentage of Grade 3 Students Achieving the Provincial Standard in Reading': 'Reading'},inplace=True)
df.rename(columns={'Percentage of Grade 3 Students Achieving the Provincial Standard in Writing': 'Writing'},inplace=True)
df.rename(columns={'Percentage of Grade 3 Students Achieving the Provincial Standard in Mathematics': 'Maths'},inplace=True)

```

```python
# Certain columns has zeros or various characters (N/R, NA, etc ...) as data, this is invalid or incomplete data
# we wil replace these with Nan or it will affect the outcome.
# We can later delete rows that that have Nan, as there are no replacement values 

zero_not_accepted = [
    'NotEnglish', 'Reading', 'Writing', 'Maths', 'LowIncomeHome', 'EducatedParents']
    
for column in zero_not_accepted:
    df[column] = df[column].replace(0.00, np.NaN)
    df[column] = df[column].replace(0.0, np.NaN)
    df[column] = df[column].replace('N/R', np.NaN)
    df[column] = df[column].replace('SP', np.NaN)
    df[column] = df[column].replace('ND', np.NaN)
    df[column] = df[column].replace('NA', np.NaN)
    df[column] = df[column].replace('N/D', np.NaN)
    df[column] = df[column].replace('S. R.', np.NaN)    
    df[column] = df[column].replace('A/D', np.NaN)    
```

```python
# Grade3 ranges
# breakdown of grades ranges into grades
grade3 = ['JK-8', 'JK-6', 'JK-5', '1-8', 'JK-3', 'K-8', 'JK-4', '1-5', 'K-6']
#grade6 = ['JK-8', 'JK-6', '6-8', '1-8', '4-8', '5-8', 'K-8', 'K-6']
#grade9 = ['9-12', 'Pre-Grde 9-12', '9', '9-10']
#grade10 = ['9-12', 'Pre-grade 9-12', '9-10']
```

```python
# Sort by grades
df = df.loc[df['Grade Range'].isin(grade3)].copy()
```

```python
# Box plot the information from describe() to help analysis

data_to_plot = (df['Reading'], df['Writing'], df['Maths'])

# Create a figure instance
fig = plt.figure(1, figsize=(9, 6))

# Create an axes instance
ax = fig.add_subplot(111)

# Add a horizontal grid to the plot, 
#
ax.yaxis.grid(True, linestyle='-', which='major', color='blue',
               alpha=0.5)

# Create the boxplot with labels
ax.set_xlabel('Grade 3')
ax.set_ylabel('Percentage of Students')
ax.set_title('Percentages of Grade 3 EQAO Passes in Reading, Writing and Maths')
bp = ax.boxplot(data_to_plot)


## add patch_artist=True option to ax.boxplot() 
## to get fill color
bp = ax.boxplot(data_to_plot, patch_artist=True)

## change outline color, fill color and linewidth of the boxes
for box in bp['boxes']:
    # change outline color
    box.set( color='#7570b3', linewidth=2)
    # change fill color
    box.set( facecolor = '#1b9e77' )

## change color and linewidth of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color='#7570b3', linewidth=2)

## change color and linewidth of the caps
for cap in bp['caps']:
    cap.set(color='#7570b3', linewidth=2)

## change color and linewidth of the medians
for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=2)

## change the style of fliers and their fill
for flier in bp['fliers']:
    flier.set(marker='o', color='#e7298a', alpha=0.5)
    
## Custom x-axis labels
ax.set_xticklabels(['Reading', 'Writing', 'Maths'])

## Remove top axes and right axes ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()

# Save the figure
fig.savefig('fig1.png', bbox_inches='tight')

```

```python
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# visualize the relationship between the features and the response using scatterplots
fig, axs = plt.subplots(1, 4, sharey=True)
ax0 = df.plot(kind='scatter', x='LowIncomeHome', y='Maths', ax=axs[0], figsize=(16, 8))
ax1 = df.plot(kind='scatter', x='EducatedParents', y='Maths', ax=axs[1])
ax2 = df.plot(kind='scatter', x='Enrolment', y='Maths', ax=axs[2])
ax3 = df.plot(kind='scatter', x='NotEnglish', y='Maths', ax=axs[3])
ax1.set_title('Percentage of Grade 3 Students Achieving the Provincial Standard in Mathematics')

## Plot line of best fit
lm = smf.ols(formula= 'Maths' + ' ~ ' + 'LowIncomeHome', data=df).fit()
regX = pd.DataFrame({'LowIncomeHome': [df.LowIncomeHome.min(), df.LowIncomeHome.max()]})
ax0.plot(regX, lm.predict(regX), c='red', linewidth=2)

lm = smf.ols(formula= 'Maths' + ' ~ ' + 'EducatedParents', data=df).fit()
regX = pd.DataFrame({'EducatedParents': [df.EducatedParents.min(), df.EducatedParents.max()]})
ax1.plot(regX, lm.predict(regX), c='red', linewidth=2)

lm = smf.ols(formula= 'Maths' + ' ~ ' + 'Enrolment', data=df).fit()
regX = pd.DataFrame({'Enrolment': [df.Enrolment.min(), df.Enrolment.max()]})
ax2.plot(regX, lm.predict(regX), c='red', linewidth=2)

lm = smf.ols(formula= 'Maths' + ' ~ ' + 'NotEnglish', data=df).fit()
regX = pd.DataFrame({'NotEnglish': [df.NotEnglish.min(), df.NotEnglish.max()]})
ax3.plot(regX, lm.predict(regX), c='red', linewidth=2)
```
