```python
# we don't need all those columns - only keep useful ones
cols = ['School Number', 'Board Name',
       'Grade Range', 'Enrolment',
       'Percentage of Students Whose First Language Is Not English',
       'Percentage of Students That Passed the Grade 10 OSSLT on Their First Attempt',
       'Percentage of Children Who Live in Low-Income Households',
       'Percentage of Students Whose Parents Have Some Unviersity Education']
    
df= df[cols]
```

```python
df.rename(columns={'Percentage of Students Whose Parents Have Some Unviersity Education': 'EducatedParents'},inplace=True)
df.rename(columns={'Percentage of Children Who Live in Low-Income Households': 'LowIncomeHome'},inplace=True)
df.rename(columns={'Percentage of Students Whose First Language Is Not English': 'NotEnglish'},inplace=True)

df.rename(columns={'Percentage of Students That Passed the Grade 10 OSSLT on Their First Attempt': 'Osslt'},inplace=True)
```

```python
# Certain columns has zeros or various characters (N/R, NA, etc ...) as data, this is invalid or incomplete data
# we wil replace these with Nan or it will affect the outcome.
# We can later delete rows that that have Nan, as there are no replacement values 

zero_not_accepted = [
    'NotEnglish', 'Osslt', 'LowIncomeHome', 'EducatedParents']
    
for column in zero_not_accepted:
    df[column] = df[column].replace(0.00, np.NaN)
    df[column] = df[column].replace(0.0, np.NaN)
    df[column] = df[column].replace('N/R', np.NaN)
    df[column] = df[column].replace('SP', np.NaN)
    df[column] = df[column].replace('ND', np.NaN)
    df[column] = df[column].replace('NA', np.NaN)
    df[column] = df[column].replace('N/D', np.NaN)
    df[column] = df[column].replace('S. R.', np.NaN)    
    df[column] = df[column].replace('A/D', np.NaN)    
```

```python
# Grade10 ranges
# breakdown of grades ranges into grades
#grade3 = ['JK-8', 'JK-6', 'JK-5', '1-8', 'JK-3', 'K-8', 'JK-4', '1-5', 'K-6']
#grade6 = ['JK-8', 'JK-6', 'JK-5', '6-8', '1-8', '4-8', '5-8', 'K-8', 'K-6']
#grade9 = ['9-12', 'Pre-Grde 9-12', '9', '9-10']
grade10 = ['9-12', 'Pre-grade 9-12', '9-10']
```

```python
# Sort by grades
df = df.loc[df['Grade Range'].isin(grade10)].copy()
```

```python
print(df.shape)
```

```python
df = df[pd.notnull(df['Osslt'])]
df = df[pd.notnull(df['LowIncomeHome'])]
df = df[pd.notnull(df['EducatedParents'])]
df = df[pd.notnull(df['NotEnglish'])]

df.head()
```

```python
print(df.shape)
```

```python
# Doing a stratified sampling by grouping by Board Name and then doing a random sample of 10 schools
# if there are not enough to sample 10, then give all the schools for that Board Name
df = df.groupby('Board Name', group_keys=False).apply(lambda x: x.sample(min(len(x), 10)))

#df = df.loc[df['Board Name']== 'Toronto DSB']
```

```python
print(df.shape)
```

```python
df.reset_index(drop=True, inplace=True)
```

```python
df.loc[:, 'Osslt'] = df['Osslt'] * 100
```

```python
# Box plot the information from describe() to help analysis

data_to_plot = (df['Osslt'])

# Create a figure instance
fig = plt.figure(1, figsize=(9, 6))

# Create an axes instance
ax = fig.add_subplot(111)

# Add a horizontal grid to the plot, 
#
ax.yaxis.grid(True, linestyle='-', which='major', color='blue',
               alpha=0.5)

# Create the boxplot with labels
ax.set_xlabel('Grade 10')
ax.set_ylabel('Percentage of Students')
ax.set_title('Percentage of Students That Passed the Grade 10 OSSLT on Their First Attempt')
bp = ax.boxplot(data_to_plot)


## add patch_artist=True option to ax.boxplot() 
## to get fill color
bp = ax.boxplot(data_to_plot, patch_artist=True)

## change outline color, fill color and linewidth of the boxes
for box in bp['boxes']:
    # change outline color
    box.set( color='#7570b3', linewidth=2)
    # change fill color
    box.set( facecolor = '#1b9e77' )

## change color and linewidth of the whiskers
for whisker in bp['whiskers']:
    whisker.set(color='#7570b3', linewidth=2)

## change color and linewidth of the caps
for cap in bp['caps']:
    cap.set(color='#7570b3', linewidth=2)

## change color and linewidth of the medians
for median in bp['medians']:
    median.set(color='#b2df8a', linewidth=2)

## change the style of fliers and their fill
for flier in bp['fliers']:
    flier.set(marker='o', color='#e7298a', alpha=0.5)
    
## Custom x-axis labels
ax.set_xticklabels(['Osslt'])

## Remove top axes and right axes ticks
ax.get_xaxis().tick_bottom()
ax.get_yaxis().tick_left()

# Save the figure
fig.savefig('fig1.png', bbox_inches='tight')

```

```python
import matplotlib.pyplot as plt
import statsmodels.formula.api as smf

# visualize the relationship between the features and the response using scatterplots
fig, axs = plt.subplots(1, 4, sharey=True)
ax0 = df.plot(kind='scatter', x='LowIncomeHome', y='Osslt', ax=axs[0], figsize=(16, 8))
ax1 = df.plot(kind='scatter', x='EducatedParents', y='Osslt', ax=axs[1])
ax2 = df.plot(kind='scatter', x='Enrolment', y='Osslt', ax=axs[2])
ax3 = df.plot(kind='scatter', x='NotEnglish', y='Osslt', ax=axs[3])
ax1.set_title('Percentage of Students That Passed the Grade 10 OSSLT on Their First Attempt')

## Plot line of best fit
lm = smf.ols(formula= 'Osslt' + ' ~ ' + 'LowIncomeHome', data=df).fit()
regX = pd.DataFrame({'LowIncomeHome': [df.LowIncomeHome.min(), df.LowIncomeHome.max()]})
ax0.plot(regX, lm.predict(regX), c='red', linewidth=2)

lm = smf.ols(formula= 'Osslt' + ' ~ ' + 'EducatedParents', data=df).fit()
regX = pd.DataFrame({'EducatedParents': [df.EducatedParents.min(), df.EducatedParents.max()]})
ax1.plot(regX, lm.predict(regX), c='red', linewidth=2)

lm = smf.ols(formula= 'Osslt' + ' ~ ' + 'Enrolment', data=df).fit()
regX = pd.DataFrame({'Enrolment': [df.Enrolment.min(), df.Enrolment.max()]})
ax2.plot(regX, lm.predict(regX), c='red', linewidth=2)

lm = smf.ols(formula= 'Osslt' + ' ~ ' + 'NotEnglish', data=df).fit()
regX = pd.DataFrame({'NotEnglish': [df.NotEnglish.min(), df.NotEnglish.max()]})
ax3.plot(regX, lm.predict(regX), c='red', linewidth=2)
```

```python
###### Extracting the Independent and Dependent Variables
#y = Dependent variable  'Percentage of Grade 9 Students Achieving the Provincial Standard in Academic Mathematics'
#X = Independent variable 'Percentage of Children Who Live in Low-Income Households'

X = df.iloc[ : , 6].values
y = df.iloc[ : , 5].values
```

```python
X=X.reshape(len(X),1)
y=y.reshape(len(y),1)
```

```python
# Splitting the dataset into the Training and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_test.size
```

```python
# Fitting Multiple Lienar Regression to the training set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train, y_train)
```

```python
# Predicting the Test Set result
y_pred = regressor.predict(X_test)
```

```python
# calculating the Coefficients
print(regressor.coef_)
```

```python
# calculating the Intercept
print(regressor.intercept_)
```

```python
# calculating the R squared value
from sklearn.metrics import r2_score
r2_score(y_test, y_pred)
```

```python
# An R-squared values of -0.303 proves the model is not a good model

# We trained our model with certain predictors using linear regression.
```

```python
#Linear Regression
import statsmodels.api as sm
```

```python
# Osslt and LowIncomeHome
X = df[['LowIncomeHome']]
Y = df['Osslt']
X = sm.add_constant(X)
model = sm.OLS(Y, X).fit()
model.summary()
```

```python
plt.figure(figsize=(10,8))
plt.plot(df['LowIncomeHome'], df['Osslt'], 'o', label='Original data', markersize=6)
intercept, slope = model.params
plt.plot(df['LowIncomeHome'], intercept + slope * df['LowIncomeHome'] , 'r-', label='Fitted Line')
plt.legend()
plt.title('Percentages Grade10 Osslt vs LowIncomeHome')
plt.ylim(0, 100)
plt.ylabel('Osslt')
plt.xlabel('LowIncomeHome')
plt.text(15, 60.0, 'y = -1.3342x + 115.4602' , fontsize=18)
plt.text(15, 55.0, 'R-Squared = 0.291' , fontsize=18)
plt.show()
```
